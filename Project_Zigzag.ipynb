```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import math as math
import pybinding as pb
from pybinding.repository import graphene
pb.pltutils.use_style()
from IPython.display import display, HTML
```

# System

## System_ImportantConstant

```python
a = 3*graphene.a_cc # lattice length.
a_cc = graphene.a_cc # carbon-carbon bond length
t = graphene.t # hopping energy of graphene.
leng_z = (7)*math.sqrt(3)*a_cc #lengh of supercell that I wanna make
m = 0.001 
plt.rcParams.update({'font.size': 15})
```

## System_Zigzag Graphene

```python
def zigzag_graphene() :
    
    lat = pb.Lattice(a1=[math.sqrt(3)*a_cc,0], a2 = [0,a]) #lattice vector
    
    lat.add_sublattices(
        ('A',[math.sqrt(3)/2*a_cc,0]),
        ('B',[math.sqrt(3)/2*a_cc+math.sqrt(3)/2*a_cc,1/2*a_cc]),
        ('C',[math.sqrt(3)/2*a_cc+math.sqrt(3)/2*a_cc,3/2*a_cc]),
        ('D',[math.sqrt(3)/2*a_cc,2*a_cc]),
        )
                        
    lat.add_hoppings(
        ([0,0], 'A', 'B', t),
        ([0,0], 'B', 'C', t),
        ([0,0], 'C', 'D', t),
        ([1,0],'B','A',t),
        ([-1,0],'D','C',t), 
        ([0,1],'D','A',t) 
    )
    
    return lat

plt.figure(figsize=(10,10))
zigzag = zigzag_graphene()
zigzag.plot()
plt.show()
```

## System_Superlattice Frame

```python
def shape1(T) : # T is the period of the shape
    rect = pb.rectangle(x=leng_z+100*m, y=10/2*math.sqrt(3)*a_cc) 
    rect_exc= pb.rectangle(x=leng_z/2, y=2*a_cc) # rectangle that wil lbe excluded

   
    tr_rect = rect.with_offset([leng_z*(T-1),a_cc])
    tr_rect_down = rect_exc.with_offset([leng_z*(T-1)+3/2*a_cc, a_cc-4*a_cc]) #3/2*a_cc
    tr_rect_up = rect_exc.with_offset([leng_z*(T-1)+3/2*a_cc, a_cc+4*a_cc]) #3/2*a_cc
    shape_1 = tr_rect - tr_rect_down -tr_rect_up    

    return shape_1


def shape2(T) : # T is the period of the shape
    rect = pb.rectangle(x=leng_z+100*m, y=10/2*math.sqrt(3)*a_cc) 
    rect_exc= pb.rectangle(x=leng_z/2, y=7*a_cc) # rectangle that wil lbe excluded

   
    tr_rect = rect.with_offset([leng_z*(T-1),a_cc])
    tr_rect_down = rect_exc.with_offset([leng_z*(T-1)+3/2*a_cc, a_cc-4*a_cc]) #3/2*a_cc
    tr_rect_up = rect_exc.with_offset([leng_z*(T-1)+3/2*a_cc, a_cc+4*a_cc]) #3/2*a_cc
    shape_1 = tr_rect -tr_rect_up    # - tr_rect_down 

    return shape_1
```

# Superlattice

## Superlattice_StructureA

```python
supercell_A = shape1(0)

print('How many supercell you want to make?')
inp = input() 

for i in range (0,int(inp)) :
    supercell_A += shape1(i)
    
    
supercell_A.plot()
#zz_gra = pb.Model(lattice, shape())#,pb.translational_symmetry(a1=18*a_cc,a2=False))
#zz_gra.plot()
#zz_gra.shape.plot()

plt.figure(figsize=(50, 20.6))
zz_gra_A = pb.Model(zigzag, supercell_A, pb.translational_symmetry(a1=(int(inp)-1)*leng_z ,a2=False))
zz_gra_A.plot()
#zz_gra.shape.plot()
```

## Superlattice_StructureB

```python
supercell_B = shape2(0)

print('How many supercell you want to make?')
inp = input() 

for i in range (0,int(inp)) :
    supercell_B += shape2(i)
    
    
supercell_B.plot()
#zz_gra = pb.Model(lattice, shape())#,pb.translational_symmetry(a1=18*a_cc,a2=False))
#zz_gra.plot()
#zz_gra.shape.plot()

plt.figure(figsize=(50, 20.6))
zz_gra_B = pb.Model(zigzag, supercell_B, pb.translational_symmetry(a1=(int(inp)-1)*leng_z ,a2=False))
zz_gra_B.plot()
#zz_gra.shape.plot()
```

## Superlattice_StructureC

```python
supercell_A = shape1(0)

print('How many supercell you want to make?')
inp = input() 

for i in range (0,int(inp)) :
    supercell_A += shape1(i)
    
    
supercell_A.plot()
#zz_gra = pb.Model(lattice, shape())#,pb.translational_symmetry(a1=18*a_cc,a2=False))
#zz_gra.plot()
#zz_gra.shape.plot()

plt.figure(figsize=(50, 20.6))
zz_gra_C = pb.Model(zigzag, supercell_A, pb.translational_symmetry(a1=(int(inp)-1)*leng_z ,a2=False))
zz_gra_C.plot()
#zz_gra.shape.plot()
```

# DATA

## Energy

```python
plt.plot(figsize=(25, 25))

solverA = pb.solver.arpack(zz_gra_A, k=20)  # for the 20 lowest energy eig_val_
solverB = pb.solver.arpack(zz_gra_B, k=20)  # for the 20 lowest energy eig_val_
solverC = pb.solver.arpack(zz_gra_C, k=20)  # for the 20 lowest energy eig_val_
eig_val_A = solverA.calc_eigenvalues()
eig_val_A.plot(label="Structure A", c='C1')
eig_val_B = solverB.calc_eigenvalues()
eig_val_B.plot(label="Structure B", c='C2')
eig_val_C = solverC.calc_eigenvalues()
eig_val_C.plot(label="Structure C", c='C4')


pb.pltutils.legend()
```

## DOS

```python
plt.plot(figsize=(25, 25))

dosA = solverA.calc_dos(energies=np.linspace(-1, 1, 200), broadening=0.05)  # [eV]
dosA.plot(label="Structure A", c='C1')

dosB = solverB.calc_dos(energies=np.linspace(-1, 1, 200), broadening=0.05)  # [eV]
dosB.plot(label="Structure B", c='C2')

dosC = solverC.calc_dos(energies=np.linspace(-1, 1, 200), broadening=0.05)  # [eV]
dosC.plot(label="Structure C", c='C4')

pb.pltutils.legend()
```

## LDOS_usingKPM

```python
kpmA = pb.kpm(zz_gra_A)
kpmB = pb.kpm(zz_gra_B)
kpmC = pb.kpm(zz_gra_C)

ldosA = kpmA.calc_ldos(energy=np.linspace(-9, 9, 2500), broadening=0.05,position=[0, 0])
ldosB = kpmB.calc_ldos(energy=np.linspace(-9, 9, 2500), broadening=0.05,position=[0, 0])
ldosC = kpmC.calc_ldos(energy=np.linspace(-9, 9, 2500), broadening=0.05,position=[0, 0])

ldosA.plot(label='A', c='C1')
ldosB.plot(label='B', c='C2')
ldosC.plot(label='C', c='C4')
    
    
plt.xlim(-2,2)

pb.pltutils.legend()
```

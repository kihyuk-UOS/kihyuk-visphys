```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import math as math
import pybinding as pb
from pybinding.repository import graphene
pb.pltutils.use_style()
```

```python
a = graphene.a_cc * math.sqrt(3) # lattice length.
a_cc = graphene.a_cc # carbon-carbon bond length
t = graphene.t # hopping energy of graphene.

```

## Armchair Graphene

```python
import math as math

def monolayer_graphene() :
    a = 0.24595 # unit cell length
    a_cc = 0.142 # carbon-carbon distance
    t = -2.8 # nearest neighbour hopping
    
    lat = pb.Lattice(a1=[3*a_cc,0], a2 = [0,math.sqrt(3)*a_cc]) #lattice vector
    
    lat.add_sublattices(
        ('A',[-5/4*a_cc,-math.sqrt(3)/4*a_cc]),
        ('B',[-1/4*a_cc,-math.sqrt(3)/4*a_cc]),
        ('C',[1/4*a_cc,math.sqrt(3)/4*a_cc]),
        ('D',[5/4*a_cc,math.sqrt(3)/4*a_cc])
        )
                        
    lat.add_hoppings(
        ([0,0], 'A', 'B', t),
        ([0,0], 'B', 'C', t),
        ([0,0], 'C', 'D', t),
        ([-1,0],'A','D',t),
        ([0,-1],'B','C',t), 
        ([1,1],'D','A',t) 
    )
    
    return lat


lattice = monolayer_graphene()
lattice.plot()
plt.show()
```

## System

```python
epsilon = 0.001

def shape() :
    rectangle = pb.rectangle(x=2*a,y=4*math.sqrt(3)*a_cc)
    hexagon = pb.regular_polygon(num_sides=6, radius=a_cc+ epsilon, angle=np.pi/6)

    trans_hexagon_down = hexagon.with_offset([a_cc, -math.sqrt(3)*2*a_cc]) #3/2*a_cc
    trans_hexagon_up = hexagon.with_offset([a_cc, math.sqrt(3)*2*a_cc]) #3/2*a_cc
    shape = rectangle - trans_hexagon_down - trans_hexagon_up

    return shape

arm_gra = pb.Model(lattice, shape(), pb.translational_symmetry(a1=a_cc,a2=False))
arm_gra.plot()
arm_gra.shape.plot()
```
